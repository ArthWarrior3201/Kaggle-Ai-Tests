```python
# Cell 2: App Script & Logo Preparation
import shutil
import os

# --- PASTE YOUR DIRECT FILE PATH HERE ---
# Example: '/kaggle/input/oraclepllogo/logo.png'
my_logo_path = '/kaggle/input/datasets/arthwarrior3201/oraclepersonalloanlogo/ChatGPT Image May 10 2026 02_49_23 PM.png' 
# ---------------------------------------

if not os.path.exists(my_logo_path) or os.path.isdir(my_logo_path):
    print("❌ ERROR: Path is invalid or a folder. Please copy the FILE path from the sidebar.")
else:
    # Prepare the environment
    shutil.copy(my_logo_path, './logo.png')
    
    with open('main.py', 'w') as f:
        f.write('''
from kivy.app import App
from kivy.uix.screenmanager import ScreenManager, Screen
from kivy.uix.image import Image
from kivy.uix.label import Label
from kivy.uix.button import Button
from kivy.uix.boxlayout import BoxLayout
from kivy.clock import Clock
from kivy.core.window import Window

# Pro Finance look: White background
Window.clearcolor = (1, 1, 1, 1)

class SplashScreen(Screen):
    def __init__(self, **kwargs):
        super().__init__(**kwargs)
        layout = BoxLayout(orientation='vertical', padding=60, spacing=20)
        # Your Golden Eye Logo
        layout.add_widget(Image(source='logo.png', size_hint=(None, None), size=(450, 450), pos_hint={'center_x': .5}))
        layout.add_widget(Label(text="Oracle Personal Loan", color=(0,0,0,1), font_size='28sp', bold=True))
        
        btn = Button(text="GET STARTED", size_hint=(0.8, None), height=70, pos_hint={'center_x': 0.5}, background_color=(0, 0.4, 0.8, 1))
        btn.bind(on_press=self.go_next)
        layout.add_widget(btn)
        self.add_widget(layout)

    def go_next(self, instance):
        self.manager.current = 'prank'

class PrankScreen(Screen):
    def __init__(self, **kwargs):
        super().__init__(**kwargs)
        self.layout = BoxLayout(orientation='vertical', padding=50)
        self.msg = Label(text="Processing KYC Documents...\\nVerifying Aadhaar & PAN", color=(0,0,0,1), font_size='18sp', halign="center")
        self.layout.add_widget(self.msg)
        self.add_widget(self.layout)

    def on_enter(self):
        # 5 second "processing" delay
        Clock.schedule_once(self.finish_prank, 5)

    def finish_prank(self, dt):
        self.msg.text = "LOAN APPROVED!\\n\\nAmount: 50,000 INR\\nArriving in 22 Minutes."
        self.msg.font_size = '24sp'
        self.msg.color = (0, 0.5, 0, 1) # Success Green

class OracleApp(App):
    def build(self):
        sm = ScreenManager()
        sm.add_widget(SplashScreen(name='login'))
        sm.add_widget(PrankScreen(name='prank'))
        return sm

if __name__ == "__main__":
    OracleApp().run()
''')
    print("✅ Logo moved and App Script written! Ready for build.")
```
